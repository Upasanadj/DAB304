```python
import pandas as pd
import datetime
from numpy import genfromtxt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
unixts = datetime.datetime.fromtimestamp(1539435366)

input_file = "Data/S1/Midterm 1/ACC.csv"
df = pd.read_csv(input_file)
new_df=df.assign(StudentNum= 1, ExamType = 2)
df_1_ACC_M1 = new_df.reset_index(drop=True)
df_1_ACC_M1.columns=['X','Y','Z','StudentNum','ExamType']
df_1_ACC_M1=df_1_ACC_M1.drop(0)
df_1_ACC_M1['time'] = pd.date_range(unixts, freq='0.03min', periods=len(df_1_ACC_M1))

df_ACC_M1 = df_1_ACC_M1

unixts = datetime.datetime.fromtimestamp(1541859555)

input_file = "Data/S1/Midterm 2/ACC.csv"
df = pd.read_csv(input_file)
new_df=df.assign(StudentNum= 1, ExamType = 3)
df_1_ACC_M1 = new_df.reset_index(drop=True)
df_1_ACC_M1.columns=['X','Y','Z','StudentNum','ExamType']
df_1_ACC_M1=df_1_ACC_M1.drop(0)
df_1_ACC_M1['time'] = pd.date_range(unixts, freq='0.03min', periods=len(df_1_ACC_M1))

df_ACC_M2 = df_1_ACC_M1

unixts = datetime.datetime.fromtimestamp(1544027337)

input_file = "Data/S1/Final/ACC.csv"
df = pd.read_csv(input_file)
new_df=df.assign(StudentNum= 1, ExamType = 1)
df_1_ACC_M1 = new_df.reset_index(drop=True)
df_1_ACC_M1.columns=['X','Y','Z','StudentNum','ExamType']
df_1_ACC_M1=df_1_ACC_M1.drop(0)
df_1_ACC_M1['time'] = pd.date_range(unixts, freq='0.03min', periods=len(df_1_ACC_M1))

df_ACC_F = df_1_ACC_M1

df_ACC1 =[]
df_ACC1 = df_ACC_M1.append(df_ACC_M2)
df_ACC1 = df_ACC1.append(df_ACC_F)
                                            
```

```python
df_ACC1.to_csv("ACC1.csv")
```
